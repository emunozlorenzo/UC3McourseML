# Introducción 

En este notebook vamos a trabajar en el uso de técnicas de aprendizaje automático basadas en análisis multivariante combinadas con técnicas de detección de outliers para detectar un posible cambio de escenario, es decir, que las observaciones del conjunto de test dejan de parecerse a las observaciones que hemos usado para entrenar el modelo.

La aplicación sobre la que vamos a trabajar es la clasificación de textos usando el conjunto de textos *benchmark* [20 newsgroups](http://qwone.com/~jason/20Newsgroups/), que está disponible para trabajar con sklearn. Los textos son *posts* escritos en inglés en grupos de noticias de diversas temáticas y la tarea consiste en adivinar el tema del posts (en qué lista de distribución se publicó) a partir del texto del mismo. 

Concretamente el desarrollo del trabajo en el notebook va a ser el siguiente. 
1. Aplicar herramientas de procesamiento de textos para conseguir representaciones numéricas de las revisiones

2. Construir un modelo de clasificaicón de textos. Dado que las representaciones numéricas de estos textos van a ser razonablemente dispersas, exploraremos tanto modelos que trabajen en el espacio original como modelos que trabajen en un espacio de dimensionalidad reducida construido con técnicas de análisis multivariante.

3. Evaluaremos las prestaciones de ese modelo usando un conjunto de test donde los textos pertenecen a las mismas categorías que las empleadas para entrenar los modelos.

4. A continuación trabajaremos la detección de novedad introduciendo en el conjunto de test textos de otras categorías distintas a las empleadas para aprender los clasificadores. Este hecho conduciría a una degradación de las prestaciones del sistema donde el clasificador de textos estuviese integrado ya que los textos de categorías no empleadas para entrenar serían clasificados forzosamente en alguna de las categorías pre-existentes.

5. Una manera de anticiparnos a la degradación de prestaciones, que en el punto 4 hemos evaluado *ex-post* sería incluir un módulo de detección de novedad en alguna de las partes del *pipeline* de procesamiento de datos empleado para la construcción del modelo. Este módulo se encargaría de detectar si los textos que llegan en el conjunto de test son lo suficientemente diferentes respecto de los empleados revisiones usadas para entrenar como para comprometer nuestra confianza en las predicciones del modelo.

# 1. Vectorización de los documentos (10%)

Aplique el pipeline estandar visto en la sesión de NLP (tokenización, homogeneización y limpieza) para el preprocesado de los textos incluidos en las listas `text_train`, `text_test` y `text_novel`.

A partir del resultado del preprocesamiento de las listas de documentos genere una representación vectorial TF-IDF para cada documento, **empleando únicamente los textos en la lista** `text_train` **para la generación de diccionario, modelo de bolsa de palabras y modelo TF-IDF**. Con estos elementos generados, transforme los documentos de las tres listas en su representación vectorial correspondiente



```python
# Libraries
import nltk
import string

# Function for Text Normalizing
def normalize(text):
    normalized_text = []
    for token in nltk.word_tokenize(text):
        token = token.lower()  # Convert all words into lowercase
        stopwords   = set(nltk.corpus.stopwords.words('english'))
        punctuation = string.punctuation
        if token not in stopwords and token not in punctuation and token.isalnum():
            lemmatizer = nltk.stem.wordnet.WordNetLemmatizer()
            token = lemmatizer.lemmatize(token)
            #snowball = nltk.stem.snowball.SnowballStemmer('english')
            #token = snowball.stem(token)
            normalized_text.append(token)
    return normalized_text


# Preprocessing all the Datasets
text_train_preproc = [normalize(x) for x in text_train]
text_test_preproc = [normalize(x) for x in text_test]
text_novel_preproc = [normalize(x) for x in text_novel]

# Sample
print(text_train[0])
print(text_train_preproc[0])
```

```python
import gensim
import matplotlib.pyplot as plt

# Create dictionary of tokens: the input is the preprocessed corpus 
D = gensim.corpora.Dictionary(text_train_preproc)
n_tokens = len(D)
print('The dictionary contains', n_tokens, 'terms')
print('First terms in the dictionary:')
for n in range(10):
    print(str(n), ':', D[n])
# Analize the word frequency per document 
plt.hist(D.dfs.values(),50)
plt.show()

# Eliminamos los términos que son números o que solo tienen una letra
bad_ids = [n for n in range(len(D)) if (D[n].isdigit()) or (D[n].isalpha() and len(D[n])==1)]
D.filter_tokens(bad_ids=bad_ids)  

# Filter extremes
# Se filtran los términos que aparecen más de un 80% en todo el corpus con el parámetro no_above
# El segundo criterio aplicado es n_keep = 1000 para quedarnos con el nº de términos indicados
# No filtramos por el umbral inferior por lo que no asignamos valor a no_below
no_above=0.8
D.filter_extremes(no_above=no_above, keep_n = 1000)

n_tokens = len(D)
print('The dictionary contains', n_tokens, 'terms')
print('First terms in the dictionary:')
for n in range(10):
    print(str(n), ':', D[n])
# Analize the word frequency per document 
plt.hist(D.dfs.values(),50)
plt.show()
```

```python
# Generar el BoW
corpus_bow_train = [D.doc2bow(doc) for doc in text_train_preproc]
corpus_bow_test = [D.doc2bow(doc) for doc in text_test_preproc]
corpus_bow_novel = [D.doc2bow(doc) for doc in text_novel_preproc]

#Limpiamos aquellas reviews que se han quedado vacías y quitamos tambien esa entrada de nuestros output
# y = [output  for review, output in zip(corpus_bow, y_train) if review ]
# corpus_bow = [review for review in corpus_bow if review]
```

```python
from gensim.models import TfidfModel
from gensim.matutils import corpus2dense, corpus2csc

# fit TFIDF model for all the corpus
model = TfidfModel(corpus_bow_train)

corpus_tfidf_train = model[corpus_bow_train]
corpus_tfidf_test = model[corpus_bow_test]
corpus_tfidf_novel = model[corpus_bow_novel]

# Dense Sparse Array
num_docs = len(corpus_bow_train)
corpus_tfidf_train_dense = corpus2dense(corpus_tfidf_train,num_terms=n_tokens,num_docs=num_docs).T
corpus_tfidf_train_sparse = corpus2csc(corpus_tfidf_train,num_terms=n_tokens,num_docs=num_docs).T
corpus_tfidf_train_array = corpus_tfidf_train_sparse.toarray()

num_docs = len(corpus_bow_test)
corpus_tfidf_test_dense = corpus2dense(corpus_tfidf_test,num_terms=n_tokens,num_docs=num_docs).T
corpus_tfidf_test_sparse = corpus2csc(corpus_tfidf_test,num_terms=n_tokens,num_docs=num_docs).T
corpus_tfidf_test_array = corpus_tfidf_test_sparse.toarray()

num_docs = len(corpus_bow_novel)
corpus_tfidf_novel_dense = corpus2dense(corpus_tfidf_novel,num_terms=n_tokens,num_docs=num_docs).T
corpus_tfidf_novel_sparse = corpus2csc(corpus_tfidf_novel,num_terms=n_tokens,num_docs=num_docs).T
corpus_tfidf_novel_array = corpus_tfidf_novel_sparse.toarray()
```

# 2. Clasificación (30%)


Entrene los siguientes modelos de clasificación de textos a partir de las representaciones vectoriales de la lista de documentos `text_train` y las etiquetas del array `y_train`:

1. LDA: *Linear Discriminant Analysis*
2. PCA+SVM: *PCA* seguido de SVM con kernel lineal. Este modelo utiliza un PCA para proyectar los datos a un espacio de dimensión reducida donde se emplea una SVM con kernel lineal como clasificación. Los hiperparámetros del modelo, que deben fijarse usando validación cruzada son:
    - Número de componentes principales de PCA
    - Valor de `C` para la SVM.
    
3. PCA+LDA: *PCA* seguido de LDA. Este modelo utiliza un PCA para proyectar los datos a un espacio de dimensión reducida donde se la clasificación se realiza con un *Linear Discriminant Analysis*. Los hiperparámetros del modelo, que deben fijarse usando validación cruzada son:
    - Número de componentes principales de PCA

## Extra (10%)

Considere el empleo de clasificadores no lineales:
- actuando directamente sobre las vectorizaciones de los documentos (sería un esquema análogo al de la primera opción, en lugar de usar LDA sobre los vectores TF-IDF, se usaría un clasificador no lineal de los que hemos visto durante el curso, kNN, Random Forest, SVM no lineal) 
- actuando sobre el espacio reducido (como en la segunda opción pero sustituyendo la SVM lineal por un clasificador no lineal).

```python
# LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score

lda = LinearDiscriminantAnalysis()
lda.fit(corpus_tfidf_train_array,y_train)
y_pred_lda = lda.predict(corpus_tfidf_test_array)
# Accuracy (con score o con accuracy_score)
print(lda.score(corpus_tfidf_test_array,y_test))
print(accuracy_score(y_pred_lda,y_test))
```

```python
# PCA + SVM
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn import svm

# Pipeline
pipe = Pipeline(steps=[('PCA',PCA()),('SVM',svm.SVC())])
# Params
param_grid = {'PCA__n_components':[1,2,4,9,16,25,36,49],
               'SVM__C':np.logspace(-3,3,10)
              }
# GridSearch
grid_pca_svm = GridSearchCV(pipe,param_grid,cv=5)
grid_pca_svm.fit(corpus_tfidf_train_array,y_train)
print(grid_pca_svm.best_params_)
print(grid_pca_svm.score(corpus_tfidf_test_array,y_test))
y_pred_pca_svm = grid_pca_svm.predict(corpus_tfidf_test_array)
```

```python
# PCA + LDA

# Pipeline
pipe = Pipeline(steps=[('PCA',PCA()),('LDA',LinearDiscriminantAnalysis())])
# Params
param_grid = {'PCA__n_components':[1,2,4,9,16,25,36,49]}
# GridSearch
grid_pca_lda = GridSearchCV(pipe,param_grid,cv=5)
grid_pca_lda.fit(corpus_tfidf_train_array,y_train)
print(grid_pca_lda.best_params_)
print(grid_pca_lda.score(corpus_tfidf_test_array,y_test))
y_pred_pca_lda = grid_pca_lda.predict(corpus_tfidf_test_array)
```

```python
# Extra 1 KNN
from sklearn.neighbors import KNeighborsClassifier
# Pipeline
pipe = Pipeline(steps=[('KNN',KNeighborsClassifier())])
# Params
param_grid = {'KNN__n_neighbors':[1,2,4,9,16,25,36]}
# GridSearch
grid_knn = GridSearchCV(pipe,param_grid,cv=5)
grid_knn.fit(corpus_tfidf_train_array,y_train)
print(grid_knn.best_params_)
print(grid_knn.score(corpus_tfidf_test_array,y_test))
y_pred_knn = grid_knn.predict(corpus_tfidf_test_array)

# Extra 1 RandomForest
from sklearn.ensemble import RandomForestClassifier
# Pipeline
pipe = Pipeline(steps=[('RF',RandomForestClassifier())])
# Params
param_grid = {'RF__n_estimators':[100,200]}
# GridSearch
grid_rf = GridSearchCV(pipe,param_grid,cv=5)
grid_rf.fit(corpus_tfidf_train_array,y_train)
print(grid_rf.best_params_)
print(grid_rf.score(corpus_tfidf_test_array,y_test))
y_pred_rf = grid_rf.predict(corpus_tfidf_test_array)

# Extra 1: SVM no lineal
```

```python
# Extra 2 PCA + KNN

# Pipeline
pipe = Pipeline(steps=[('PCA',PCA()),('KNN',KNeighborsClassifier())])
# Params
param_grid = {'PCA__n_components':[1,2,4,9,16,25,36,49],
              'KNN__n_neighbors':[1,2,4,9,16,25,36]
              }
# GridSearch
grid_pca_knn = GridSearchCV(pipe,param_grid,cv=5)
grid_pca_knn.fit(corpus_tfidf_train_array,y_train)
print(grid_pca_knn.best_params_)
print(grid_pca_knn.score(corpus_tfidf_test_array,y_test))
y_pred_pca_knn = grid_pca_knn.predict(corpus_tfidf_test_array)

# Extra 2 PCA + RandomForest

# Pipeline
pipe = Pipeline(steps=[('PCA',PCA()),('RF',RandomForestClassifier())])
# Params
param_grid = {'PCA__n_components':[1,2,4,9,16,25,36,49],
              'RF__n_estimators':[100,200]
              }
# GridSearch
grid_pca_rf = GridSearchCV(pipe,param_grid,cv=5)
grid_pca_rf.fit(corpus_tfidf_train_array,y_train)
print(grid_pca_rf.best_params_)
print(grid_pca_rf.score(corpus_tfidf_test_array,y_test))
y_pred_pca_rf = grid_pca_rf.predict(corpus_tfidf_test_array)
```

# 5. Detección de novedad (40%)

Vamos a emplear los métodos de detección de *outliers* vistos en la sesión correspondiente para intentar identificar la presencia de los documentos de la lista `text_novel`. Considere **al menos dos métodos** de detección de *outliers* para el estudio.

El empleo de espacios de dimensión reducida nos abre una opción extra para aplicar la detección de novedad. Concretamente podemos plantear la detección de novedad en estos 3 momentos:
 
1. En el espacio de entrada original, con los vectores de TF-IDF
2. En el espacio de dimensión reducida creado por el clasificador LDA que actúa directamente sobre los vectores TF-IDF (el del primer modelo de clasificación)
3. En el espacio de dimensión reducida a la salida del PCA en el clasificador PCA+SVM o del modelo PCA+LDA, si es que cada modelo efectuase un PCA con distinto número de componentes principales.

Evalúe la idoneidad de aplicar la detección de novedad en cada uno de estos escenarios y para distintos valores del parámetro de contaminación de los detectores de *outliers* que emplee (al menos dos detectores).

La manera de plantear el experimento debe ser:
- **entrenar el detector de outliers únicamente con las observaciones de entrenamiento** (vectorizaciones que salgan textos de la lista `text_train`). Estas observaciones deberán haber sido transformadas al espacio reducido correspondiente (si procede).
- **evaluar** (`predict`) con las vectorizaciones resultantes de concatenar las listas de documentos `text_test` y `text_novel` y calcular las tasas de detección de outliers.
- calcular las **matrices de confusión** donde los *outliers* son la clase real $-1$ y los datos de test cuyas clases sí se usaron para entrenar la clase real $+1$. Las clases predichas las proporciona el método `predict` (punto anterior).

**Discuta cuál es la mejor estrategia para identificar los outliers**, comentando tanto la aparición de **falsos positivos** (outliers que parecen de clases presentes en el conjunto de entrenamiento) como de **falsos negativos** (textos del conjunto de test pertenecientes a clases que sí estaban presentes en el entrenamiento pero que el detector de novedad etiqueta como *outliers*).
